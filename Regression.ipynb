```python
#Simple Linear Regression + Plot

import numpy as np
import matplotlib.pyplot as plt

# Given data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Calculate mean of X and Y
x_mean = np.mean(X)
y_mean = np.mean(Y)

# Calculate slope (b1) and intercept (b0)
b1 = np.sum((X - x_mean) * (Y - y_mean)) / np.sum((X - x_mean) ** 2)
b0 = y_mean - b1 * x_mean

# Predicted values
Y_pred = b0 + b1 * X

# Print regression equation
print("Intercept (b0):", b0)
print("Slope (b1):", b1)
print("Regression Equation: Y =", b0, "+", b1, "* X")

# Plot the data points and regression line
plt.scatter(X, Y, label="Data Points")
plt.plot(X, Y_pred, label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent and dependent variables
X = df[["Area", "Rooms"]]
y = df["Price"]

# Add constant (intercept)
X_const = sm.add_constant(X)

# Fit Multiple Linear Regression
model = sm.OLS(y, X_const).fit()

# Regression summary
print("Multiple Linear Regression Summary:\n")
print(model.summary())

# Calculate VIF
vif = pd.DataFrame()
vif["Feature"] = X_const.columns
vif["VIF"] = [variance_inflation_factor(X_const.values, i)
              for i in range(X_const.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Fit a 2nd-degree polynomial
coefficients = np.polyfit(X, Y, 2)

# Create polynomial model
polynomial_model = np.poly1d(coefficients)

# Generate smooth curve
X_smooth = np.linspace(min(X), max(X), 100)
Y_smooth = polynomial_model(X_smooth)

# Print polynomial equation
print("2nd-degree Polynomial Equation:")
print(polynomial_model)

# Plot data points and polynomial curve
plt.scatter(X, Y, label="Data Points")
plt.plot(X_smooth, Y_smooth, label="Polynomial Curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([10, 20, 30, 40, 50])
Y = np.array([15, 35, 40, 50, 65])

# Fit Simple Linear Regression
coeffs = np.polyfit(X, Y, 1)
model = np.poly1d(coeffs)

# Predictions
Y_pred = model(X)

# Residuals
residuals = Y - Y_pred

# Print regression equation
print("Regression Equation:")
print(model)

# Residuals plot
plt.figure()
plt.scatter(X, residuals)
plt.axhline(0, linestyle='--')
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot (Heteroscedasticity Check)")
plt.show()
```
